**-We can see that the release_date, revenue and the budget are objects so let's change each of them into there appropriate type**

```python
movies.release_date = pd.to_datetime(movies['release_date'],  errors='coerce')
movies.revenue = pd.to_numeric(movies.revenue,errors='coerce',downcast="float")
movies.budget  = pd.to_numeric(movies.budget ,errors='coerce',downcast="float")
movies.popularity  = pd.to_numeric(movies.popularity ,errors='coerce',downcast="float")

```

```python
movies.info()
```

```python
movies.set_index('id',inplace = True)
movies.head(3)
```

```python
movies.describe()
```

**-We can see here that in the budget and revenue have  0's  it should be nulls to make the analysis easier**

```python
movies.revenue.replace(0,np.nan,inplace = True)
movies.budget.replace(0,np.nan,inplace = True)

```

```python
movies.describe()
```




> **Handling duplicates**


---



---





#### **We will focus on the movies that have more than 160 vote as this is the median of the voting count and because most of the analysis will done to the revenue and numeric columns i will drop all the null value in these columns(We can't estimate the revenue or the budget so it's better to drop them)**

```python
movies.dropna(subset=['revenue','budget'],inplace = True)
movies = movies.loc[movies.vote_count>= 160]
```

```python
movies.info()
```

```python
movies.reset_index(inplace= True)
movies.drop_duplicates(subset=['id'], inplace = True)
movies.set_index(['id'],inplace = True)
```

```python
movies.info()
```

**-make the revenue and the budget by millions**

```python
movies.revenue = movies.revenue.div(1000000)
movies.budget = movies.budget.div(1000000)
movies.head(3)
```

**-Now let's handle the non_numeric columns**

```python
movies.head(3)
```

```python
movies.belongs_to_collection = movies.belongs_to_collection.apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else np.nan )
```

```python
movies.belongs_to_collection = movies.belongs_to_collection.apply(lambda x: x['name'] if isinstance(x,dict) else np.nan)
movies.belongs_to_collection
```

```python
movies.genres = movies.genres.apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else np.nan )
```

```python
movies.genres = movies.genres.apply(lambda x:"|".join(i['name'] for i in x))
movies.genres
```

```python
movies.production_companies = movies.production_companies.apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else np.nan )
```

```python
movies.production_companies = movies.production_companies.apply(lambda x:"|".join(i['name'] for i in x))
movies.production_companies
```

## Find your next Movie

3. __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
mask_genres = movies['genres'].str.contains('Action') | movies['genres'].str.contains('Science Fiction')
mask_actor = movies.Actors.str.contains("Bruce Willis")
```

```python
movies.loc[mask_actor & mask_genres,['title','belongs_to_collection','vote_average']].sort_values(by = "vote_average", ascending = False).set_index('title')
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
mask_director = movies['director'].str.contains('Quentin Tarantino') 
mask_actor = movies.Actors.str.contains("Uma Thurman")
```

```python
movies.loc[mask_actor & mask_director,['title','belongs_to_collection','vote_average']].sort_values(by = "vote_average", ascending = False).set_index('title')
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
mask_company = movies['production_companies'].str.contains('Pixar').fillna(False) 
mask_date = movies.release_date.between("2010-01-01", "2015-12-31")
```

```python
movies.loc[mask_company & mask_date,['title','belongs_to_collection','vote_average']].sort_values(by = "vote_average", ascending = False).set_index('title')
```

__Search 4: Action or Thriller Movie with  minimum Rating of 7.5 (most recent movies first)__

```python
mask_genres = movies['genres'].str.contains('Action') | movies['genres'].str.contains('Thriller')
mask_actor = movies.vote_average >= 7.5
```

```python
movies.loc[mask_genres & mask_actor,['title','belongs_to_collection','vote_average','release_date']].sort_values(by = "release_date", ascending = False).set_index('title')
```
